# isnull() i notnull()
Podczas tworzenia ramki danych z pliku CSV, wiele pustych kolumn jest importowanych jako wartości null do ramki danych, co później stwarza problemy podczas operacji na tej ramce danych.

Czasami sam nasz zbiór danych zawiera brakujące punkty danych, które po zaimportowaniu do ramki danych zostaną przekonwertowane na wartości null lub NaN.

Metody `isnull()` i `notnull()` w Pandas służą do sprawdzania i zarządzania wartościami NULL w ramce danych.

```python
df.isnull()
```

`isnull()` zwróci True, jeśli w wartości znajduje się jakakolwiek wartość NaN, w przeciwnym razie zwróci False.

Jednak przeglądanie tak dużego zbioru danych z wartościami True lub False byłoby bardzo żmudne. Aby to uprościć, możemy zagregować te dane i sprawdzić, czy istnieją jakiekolwiek wartości null, używając `sum()` na wyniku `isnull()`, jak poniżej.

```python
df.isnull().sum()
```

Z metody `counts()` już wiedzieliśmy, że nasz zbiór danych nie ma wartości null, co potwierdziliśmy również za pomocą metody `isnull().sum()`.

Przeciwieństwem `isnull()` jest metoda `notnull()`, która zwróci True dla każdej wartości NIE-NULL i False dla wartości null.

Tutaj również możemy użyć funkcji agregującej `.sum()`, aby uzyskać liczbę.

```python
df.notnull()
```

```python
df.notnull().sum()
```

# value_counts()

Pandas `value_count()` to kolejna metoda służąca do pokazywania wszystkich unikalnych wartości wraz z ich liczbą dla danej kolumny. Ta metoda jest stosowana tylko do pojedynczych kolumn.

Wynikowy obiekt będzie posortowany w porządku malejącym, tak aby pierwszy element był najczęściej występującym elementem. Domyślnie wyklucza wartości NA.

```python
df['Weather'].value_counts()
```

Powyżej wspomniałem, że `value_counts()` stosuje się do pojedynczej kolumny, ale co, jeśli mamy wymóg uzyskania liczby unikalnych wartości dla podzbioru kolumn lub całej ramki danych.

Istnieją sposoby, dzięki którym możemy rozwiązać ten problem.
* Pętla po każdej kolumnie.
* Użycie `apply` i przekazanie `pd.Series.value_counts`
* itd.

```python
df[['Wind Speed_km/h', 'Weather']].astype(str).apply(pd.Series.value_counts)
```

# describe()
Ostatnią, ale nie mniej ważną, jest metoda `describe()` w Pandas, która służy do przeglądania podstawowych szczegółów statystycznych, takich jak percentyle, średnia, odchylenie standardowe itp., dla ramki danych lub serii wartości numerycznych.

Pamiętaj, że metoda `describe()` domyślnie zwraca szczegóły dla wartości numerycznych, jednak możemy również sprawdzić szczegóły dla typu `object`. W takim przypadku metoda zwróci inny wynik, który obejmuje całkowitą liczbę wartości, liczbę unikalnych wartości, najczęściej występującą wartość oraz częstotliwość jej występowania dla każdej kolumny typu `object`.

```python
df.describe()
```

Jeśli przyjrzymy się uważnie wynikowi, łatwo zauważymy, że kolumny `Weather` i `Date/Time` nie są wyświetlane, ponieważ ich typem danych jest `object`.

Wyświetlmy je teraz.

```python
df.describe(include=['object'])
```

Jeśli chcesz zobaczyć statystyki dla wszystkich kolumn, uwzględniając wszystkie typy danych, użyj poniższego kodu.

```python
df.describe(include=['object', 'int', 'float'])
```

Metoda `describe()` ma dwa dodatkowe parametry oprócz `include`, a mianowicie `percentile` i `exclude`. Polecam je sprawdzić i zgłębić.
